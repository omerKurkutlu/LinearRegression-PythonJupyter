```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([10,9,2,15,10,16,11,16])    # Set the input 
y = np.array([95,80,10,50,45,98,38,93])
```

```python
x = x.reshape(-1,1)
```

```python
linReg = LinearRegression()        # Creat the model
linReg.fit(x,y)                    # Train the model  
prediction = linReg.predict(x)     # predict y values depends on x values
```

```python
plt.scatter(x,y,color="green",s=30)    # put x&y values in graph
plt.plot(x,prediction,color="red")     # plot the Linear regression
plt.title("Linear Regression")     # Set Title of Graph
plt.xlabel("Number of hours spent driving")               # Set name of x- axis
plt.ylabel("Risk Score")           # Set name of y- axis
plt.show()                         #Display the graph
```
